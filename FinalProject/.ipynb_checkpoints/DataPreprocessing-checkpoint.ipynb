```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import tensorflow as tf
import numpy as np
import scipy as sp
import sklearn as sk
```

```python
PROJECT_PATH = '/pine/scr/s/i/siyangj/DeepStock/FinalProject/'
DATA_PATH = os.path.join(PROJECT_PATH,'Data/')
DATA_HDF5 = os.path.join(DATA_PATH,'clean.h5')
DATA_HDF5_ID = 'CLEAN_DATA'

RETURN_DATA = os.path.join(DATA_PATH,'return.h5')
RETURN_ID = 'RETURN'
STOCK_NAMES = ['AAPL','MSFT','AMZN','GOOG','BRKB',
              'FB','JNJ','JPM','XOM','V','WMT','BAC',
              'UNH','PFE','WFC','VZ','PG','CVX','T',
              'INTC','CSCO','HD','MA','KO','BA',
              'MRK','ORCL','DIS','CMCSA','PEP','C',
              'MCD','PM','DWDP','ABBV','NFLX','MDT',
              'ABT','NVDA','AMGN','NKE','ADBE','MO','LLY',
              'MMM','IBM','HON','UNP','ACN',
              'UTX']

VOLUME_NORMALIZATION_TABLE = os.path.join(DATA_PATH,'vol_norm.h5')
VOLUME_NORMALIZATION_ID = 'VOLUME_NORMALIZATION'

DATA_PREPROCESSED_1 = os.path.join(DATA_PATH,'preprocessed_1.h5')
DATA_PREPROCESSED_1_ID = 'DATA_PREPROCESSED_1'


NUM_DAYS = 5
NUM_PER_DAY = 390
NUM_STOCKS = 50
NUM_PER_STOCK = 5
```

```python
df = pd.read_hdf(DATA_HDF5,DATA_HDF5_ID)
```

```python
## Convert each day
def convert_to_return_each_day(y):
    y1 = np.array(y[:-1,:])
    y2 = np.array(y[1:,:])
    y[0,:] = 0
    y[1:,:] = (y2-y1)/y1
    return y
```

```python
non_vol = []
for i in range(NUM_STOCKS * NUM_PER_STOCK):
    if i%NUM_PER_STOCK!=4:
        non_vol.append(i)
vol = []
for i in range(NUM_STOCKS * NUM_PER_STOCK):
    if i%NUM_PER_STOCK==4:
        vol.append(i)
```

```python
def convert_to_return():
    for i in range(NUM_DAYS):
        #print(i)
        #print(i*NUM_PER_DAY,(i+1)*NUM_PER_DAY)
        df.iloc[i*NUM_PER_DAY:(i+1)*NUM_PER_DAY,non_vol] = \
        convert_to_return_each_day(df.iloc[i*NUM_PER_DAY:(i+1)*NUM_PER_DAY,non_vol].values)
        #print(i)
```

```python
convert_to_return()
```

```python
## Drop the start of each day
day_starts = np.arange(0,NUM_DAYS)*NUM_PER_DAY
df.index[day_starts]
```

```python
df.drop(df.index[day_starts],inplace=True)
```

```python
df.to_hdf(RETURN_DATA,RETURN_ID)
########################################
## Finished converting to returns
```

```python
df = pd.read_hdf(RETURN_DATA,RETURN_ID)
```

```python
df
```

```python
def scale_volume():
    y = df.iloc[:,vol].values
    m = y.mean(axis=0)
    v = y.var(axis=0)
    df.iloc[:,vol] = (y-m)/v
    return m,v
```

```python
dfm, dfv = scale_volume()
```

```python
vol_norm_data = {'Stock':STOCK_NAMES,'Mean':dfm,'Variance':dfv}
VOLUME_NORMALIZATION_PARAMS = pd.DataFrame(data=vol_norm_data)
```

```python
VOLUME_NORMALIZATION_PARAMS.set_index('Stock',inplace=True)
```

```python
VOLUME_NORMALIZATION_PARAMS.to_hdf(VOLUME_NORMALIZATION_TABLE,VOLUME_NORMALIZATION_ID)
```

```python
df.to_hdf(DATA_PREPROCESSED_1,DATA_PREPROCESSED_1_ID)
VOLUME_NORMALIZATION_PARAMS.to_hdf(DATA_PREPROCESSED_1,VOLUME_NORMALIZATION_ID)
##############################################################
## Finished normalizing volume first time
```
